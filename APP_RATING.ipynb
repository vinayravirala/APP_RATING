```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

# Generate sample data
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

# Initialize the AdaBoost classifier
adaboost_classifier = AdaBoostClassifier()

# Train the AdaBoost classifier
adaboost_classifier.fit(X_train, y_train)

# Make predictions on the test set
adaboost_pred_train = adaboost_classifier.predict(X_train)
adaboost_pred_test = adaboost_classifier.predict(X_test)

# Calculate accuracy for AdaBoost classifier
adaboost_train_accuracy = accuracy_score(y_train, adaboost_pred_train)
adaboost_test_accuracy = accuracy_score(y_test, adaboost_pred_test)

# Create a DataFrame to store metrics
data = {'model': ['AdaBoost Classifier', 'AdaBoost Classifier'],
        'dataset': ['train', 'test'],
        'accuracy %': [adaboost_train_accuracy * 100, adaboost_test_accuracy * 100]}

df_metrics_clf = pd.DataFrame(data)

# Plot the accuracy comparison using seaborn
g = sns.catplot(col='dataset', data=df_metrics_clf, kind='bar', x='model', y='accuracy %')

# Adding annotations to bars
for ax in g.axes.ravel():
    for c in ax.containers:
        ax.bar_label(c, label_type='edge')
    ax.margins(y=0.2)

plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, VotingClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score



# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=20)

# Initialize individual classifiers
gb_clf = GradientBoostingClassifier(random_state=20)
ab_clf = AdaBoostClassifier(random_state=20)

# Hyperparameter tuning using GridSearchCV for Gradient Boosting
param_grid_gb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.1],
    'max_depth': [3, 4]
}
grid_search_gb = GridSearchCV(gb_clf, param_grid_gb, cv=5)
grid_search_gb.fit(X_train, y_train)
best_gb_clf = grid_search_gb.best_estimator_

# Hyperparameter tuning using GridSearchCV for AdaBoost
param_grid_ab = {
    'n_estimators': [100, 200],
    'learning_rate': [0.1]
}
grid_search_ab = GridSearchCV(ab_clf, param_grid_ab, cv=5)
grid_search_ab.fit(X_train, y_train)
best_ab_clf = grid_search_ab.best_estimator_

# Train individual classifiers with best hyperparameters
best_gb_clf.fit(X_train, y_train)
best_ab_clf.fit(X_train, y_train)

# Create a voting ensemble classifier with hard voting
voting_clf = VotingClassifier(
    estimators=[('gb', best_gb_clf), ('ab', best_ab_clf)],
    voting='hard'
)

# Train the voting ensemble classifier
voting_clf.fit(X_train, y_train)

# Make predictions on the test set with ensemble
y_pred = voting_clf.predict(X_test)

# Make predictions on test set with individual classifiers
y_pred_gb = best_gb_clf.predict(X_test)
y_pred_ab = best_ab_clf.predict(X_test)

# Calculate accuracies
accuracy_ensemble = accuracy_score(y_test, y_pred)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
accuracy_ab = accuracy_score(y_test, y_pred_ab)

# Prepare data for plotting (classifier names and accuracies)
classifier_names = ['Gradient Boost', 'AdaBoost', 'Ensemble']
accuracies = [accuracy_gb, accuracy_ab, accuracy_ensemble]

# Plot the performance comparison of individual classifiers and the ensemble
plt.figure(figsize=(6, 4))
plt.bar(classifier_names, accuracies, color=['green', 'red', 'black'])
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.title('Performance of Individual Classifiers vs. Ensemble')
plt.tight_layout()
plt.show()


```
